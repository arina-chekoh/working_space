```python
import requests
import csv 
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dataforthai.com/business/objective/10801'

source = requests.get(url)
soup = BeautifulSoup(source.content, 'lxml', from_encoding='utf-8')
```

```python
div = soup.find('div', id="divResult")
print(div.prettify())
```

```python
table = div.find('table', class_='datatable')
print(table.prettify())
```

```python
rows = table.find_all('tr')

data = [] # list to store text of all elements in all the rows.
for i, row in enumerate(rows):
    # Find all elements in this row.
    # Each element is enclosed by either <th> or <td> tag.
    if i==0:
        elements = row.find_all('th')
    else:
        elements = row.find_all('td')
    
    # list to store text of all the elements in this row:
    data_elements_in_row = []
    
    for element in elements:
        text = element.text
        data_elements_in_row += [text]
    
    data += [data_elements_in_row]
```

```python
data
```

```python
cols = data[0]
rows = data[1:]
with open('locate.csv', 'w', encoding="utf-8-sig")  as f:
    write = csv.writer(f)
    write.writerow(cols)
    write.writerows(rows)
```
