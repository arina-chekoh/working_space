#### read stored data from DataPlatform

```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

```python
dir_raw_id = 76
dir_process_id = 77
```

```python
datanode = DataNode()
```

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='results.csv'))
meta
```

```python
data = pd.read_csv('./results.csv')
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Rating'] = df['Rating'].str.replace("out of 5 stars", "")
```

```python
df = df.rename(columns = {'Description' : 'Title'})
```

#### write DataNode to DataPlatform

```python
datanode.write(df=df, directory=dir_process_id, name="amazon_scrape", profiling=True, lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='results.csv')])
```
